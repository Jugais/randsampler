```python
import pandas as pd
import numpy as np
```

```python
train = pd.read_csv('virtual_resin.csv', index_col=0)
```

```python
from src.sampler import RandomSampler as rs
```

```python
sampler = rs.setup(train.values, batch_size=10)
```

```python
sampler.set_constraints('sum', value=1, 
    cols=[2,3,4],
    max_used=3)
sampler.set_constraints('func', fn=lambda x: (0<x[0]<1) and (0<x[1]<1))
sampler.set_constraints('categories', col=5, values = train.temperature.unique())

result = sampler.sample(100)
```
